```python
import requests
from BeautifulSoup import BeautifulSoup
import mechanize
import logging
import sys
from time import time
```

```python
block_and_lot = {'06165': {'00{}'.format(i):None for i in range(43,73)},
             '06176' : {'00{}'.format(i):None for i in range(12,37)}}
```

```python
times = []

for block in block_and_lot:
    
    for lot in block_and_lot[block]:
        
        t0 = time()
        br = mechanize.Browser()
        site = 'http://nycserv.nyc.gov/NYCServWeb/WaterChargeHistorySearch.jsp'
        br.open(site)
        br.select_form(name="NycservProtocolForm")

        # Set form data
        br.form['WATER_PAYMENT_HISTORY_ACC_NUMBER'] = ''
        br.form['BOROUGH'] = ['3']
        br.form['BLOCK'] = block
        br.form['LOT'] = lot

        br.form.find_control("ServiceName").readonly = False
        br.form.find_control("LinkType").readonly = False
        br.form.find_control("NycservRequest").readonly = False

        br.form['ServiceName'] = 'DEP_WATER_PAYMENT_HISTORY_SEARCH_BY_BBL'
        br.form['LinkType'] = 'B'
        br.form['NycservRequest'] = 'ChannelType=ct/Browser|RequestType=rt/Business|SubSystemType=st/Payments|AgencyType=at/ECB|ServiceName=DEP_WATER_PAYMENT_HISTORY_SEARCH_BY_BBL|MethodName=NONE|ParamCount=undefined|WATER_PAYMENT_HISTORY_ACC_NUMBER=|BOROUGH=3|BLOCK={}|LOT={}|PageID=WaterChargeHistorySearch|LinkType=B'.format(block,lot)

        # Submit form
        form_submit = br.submit()

        # Read in the HTML
        get_acct_number_html = form_submit.read()
        soup = BeautifulSoup(get_acct_number_html)

        try:
            inputs = soup.findAll('input')
            
            name = [tag.get('value') for tag in inputs if tag['name'] == 'GRID_NAME_0'][0]
            address = [tag.get('value') for tag in inputs if tag['name'] == 'GRID_ADDRESS_0'][0]
            accnt_number = [tag.get('value') for tag in inputs if tag['name'] == 'GRID_ACCT_NUM_0'][0]
            block_and_lot[block][lot] = (name, address, accnt_number)
        except Exception as e:
            print 'Error: {}'.format(e)
            pass
        
        t1 = time()
        
        times.append(t1-t0)
        
        print 'Time for request: {}'.format(t1-t0)
        print 'Average time per request: {}'.format( sum(times)/float(len(times)))
        print block_and_lot
```

```python
with open('account_numbers.txt', 'w') as f:
    for block in block_and_lot:
        for lot in block_and_lot[block]:
            f.write('{} {} {}\n'.format(block, lot, block_and_lot[block][lot]))
```
